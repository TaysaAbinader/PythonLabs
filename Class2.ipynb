

Question 2. Using the linearly spaced numerical array of x from 0 to 10 having 100 samples, draw a graph of the function:
 y = (x-2)(x-8).

Annotate the graph:

Add grid lines to the graph.

Add label for the x axis: "x-axis".

Add label for the y axis: "y-axis".

Add title for the graph: "Polynomial function".


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)
y = (x-2)*(x-8)

fig, ax = plt.subplots (figsize = (10,15))
plt.plot(x, y)
plt.grid(True)
plt.xlabel('x-axis', fontsize = 14)
plt.ylabel('y-axis', fontsize = 14)
plt.title('Polynomial function', fontsize = 16)
plt.show()
```

Question 3. Using linearly spaced numerical array of x from 0 to 10 having 100 samples draw in the same graph two polynomials:

y1 = (x-2)*(x-8)

y2 = (x-2)*(x-5)*(x-8)

Add grid-lines to the graph.


```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots ()

x = np.linspace(0,10,100)
y1 = (x-2)*(x-8)
y2 = (x-2)*(x-5)*(x-8)

plt.plot(x, y1)
plt.plot(x, y2)
plt.grid(True)
plt.xlabel('x-axis', fontsize = 14)
plt.ylabel('y-axis', fontsize = 14)
plt.title('Polynomial function', fontsize = 16)
plt.show()
```

Question 4. Using linearly spaced numerical array of x from 0 to 10 having 100 samples draw in the same graph two polynomials:

y1 = (x-2)*(x-8)

y2 = (x-2)*(x-5)*(x-8)

Annotate the graph:

Add labels 'y1' and 'y2' to the plots.

Add grid lines to the graph.

Add x-label: 'x-axis'.

Add y-label: 'y-axis'.

Add title: 'Polynomial functions'.

Add legend to the the graph.




```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots ()

x = np.linspace(0,10,100)
y1 = (x-2)*(x-8)
y2 = (x-2)*(x-5)*(x-8)

plt.plot(x, y1, label = 'y1')
plt.plot(x, y2, label = 'y2')
plt.grid(True)
plt.xlabel('x-axis', fontsize = 14)
plt.ylabel('y-axis', fontsize = 14)
plt.title('Polynomial functions', fontsize = 16)
plt.legend()
plt.show()
```

Question 5. Using linearly spaced numerical array of x from 0 to 10 having 100 samples draw in the same graph two polynomials:

y1 = (x-2)*(x-8)

y2 = (x-2)*(x-5)*(x-8)

Using two subplots, plot the first polynomial on the first axes and then the second polynomial on the second axes. 

Use figure size of (6, 8).

Annotate the graph:

Add grid lines to both plots.

Add title: 'y1' to the first subplot.

Add title: 'y2' to the second subplot.

Add x-label: 'x-axis' to the second plot.
 


```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots (2, figsize = (6,8))

x = np.linspace(0,10,100)
y1 = (x-2)*(x-8)

plt.sca(ax[0])
plt.plot(x, y1, label = 'y1')
plt.grid(True)
plt.title('y1')


y2 = (x-2)*(x-5)*(x-8)
plt.sca(ax[1])
plt.plot(x, y2, label = 'y2')
plt.grid(True)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('y2')
plt.show()
```

Question 6. Calculate values for two functions f(x) = np.sqrt(2*x) and g(x) = 5 - x.  Use linearly spaced numerical array of x from 0 to 10 having 100 samples.

Use two subplots and figure size of (6, 8).

Draw the full graphs of both functions into the first axes. This part of the graph shows the full graphs.

Draw the full graphs of both functions also into the second axes. This part is used for zooming in to the details.

Add grid lines to both subplots.

Change the x-axis and y-axis limits for the second axes, so that you can find graphically the intersection point (x0, y0) for the two equations (e.g. when f(x) = g(x)). Try to get the value minimum three significant figures accuracy.

Read the intersection point (x0, y0) from the graph and draw it to the second graph using red circle. 



```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots (2, figsize = (6,8))

x = np.linspace(0,10,100)
y1 = np.sqrt(x*2)
y2 = 5 -x


plt.sca(ax[0])
plt.plot(x, y1, label = 'y1')
plt.plot(x, y2, label = 'y2')
plt.grid(True)
plt.xlabel('x-axis')
plt.ylabel('y-axis')


plt.sca(ax[1])
plt.plot(x, y1, label = 'y1')
plt.plot(x, y2, label = 'y2')
plt.grid(True)
plt.xlim(2.682,2.685)
plt.ylim(2.315,2.318)
plt.xlabel('x-axis')
plt.ylabel('y-axis')

x0 = 2.68345
y0 = 2.31655
plt.plot(x0,y0, 'ro')
plt.show()
```
